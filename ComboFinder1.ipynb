ComboFinder
Find all the list of products whose sum-of-price is between 290 and 310.

ProductList = {p1:10, p2:15, p3:20, p4:25, p5:30, p6:35, p7:50}

```python
import random as r
Product_L = {'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35, 'p7':50,
               'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70,
               'p14':45}
Result_List  = set()
Iterations  = 1000
for i in range(Iterations):
    SetSize = r.randint(2, len(Product_L)-1)
    ComboList = r.sample(list(Product_L.keys()),SetSize)
    ComboList.sort()
    ComboSum = sum([ Product_L[i] for i in ComboList])
    if ComboSum>= 290 and ComboSum<= 3310:
      Result_List.add(tuple(ComboList))
for r in Result_List:
	print (r)
print ("\nTotal Sets--> ", len(Result_List), "\n")
```
